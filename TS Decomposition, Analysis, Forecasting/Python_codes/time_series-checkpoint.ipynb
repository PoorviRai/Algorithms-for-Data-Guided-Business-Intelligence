## Load Dataset

````python
data=pd.read_csv("./Amtrak_data.csv")
data=data[~np.isnan(data["Ridership"])]
data.index=pd.Index(sm.tsa.datetools.dates_from_range("1991m1","2004m3"))
del data["Month"]
````

Plot the time series data

````python
plt.plot(data)
plt.show()
````

Lagged scatterplots from order 1 to 9

````python
for i in range(1,10):
    plt.subplot(3,3,i)
    pd.plotting.lag_plot(data,lag=i)
````

## Autocorrelation Function

````python
sm.graphics.tsa.plot_acf(data,lags=12)
````

Plot the difference of TS

````python
plt.plot(np.diff(data["Ridership"],n=1))
plt.show()
````

## Seasonal and Trending Decomposition

````python
stl=seasonal.seasonal_decompose(data)
ax=plt.subplot(411)
ax.set_title("data")
plt.plot(stl.observed)
ax=plt.subplot(412)
ax.set_title("residual")
plt.plot(stl.resid)
ax=plt.subplot(413)
ax.set_title("seasonal")
plt.plot(stl.seasonal)
ax=plt.subplot(414)
ax.set_title("trend")
plt.plot(stl.trend)

plt.show()
````

The trend of time series

````python
plt.plot(data)
plt.plot(stl.trend)

plt.show()
````

Seasonality of the time series

````python
plt.plot(stl.seasonal)
plt.show()
````

Two types of moving averages

````python
ma_trailing=data.rolling(12,center=False).mean()
ma_centered=data.rolling(12,center=True).mean()

_,ax=plt.subplots()
ax.plot(data,label="Ridership")
ax.plot(ma_trailing,label="Trailing Moving Average")
ax.plot(ma_centered,label="Centered Moving Average")
legend=ax.legend(loc=0)
````

## Box Tests

````python
data=pd.read_csv("./Amtrak_data.csv")
data=data[~np.isnan(data["Ridership"])]
data.index=pd.Index(sm.tsa.datetools.dates_from_range("1991m1","2004m3"))
del data["Month"]
````

````python
stl=seasonal.seasonal_decompose(data)
````

````python
dir(stl)
````

````python
residuals=stl.resid
residuals[0:10]
````

````python
residuals.tail()
````

````python
residuals.dropna()[0:5]
````

+ __Box-Ljung Test:__

````python
BoxLjung=sm.stats.diagnostic.acorr_ljungbox(residuals.dropna(), boxpierce=False)
````

````python
help(sm.stats.diagnostic.acorr_ljungbox)
````

````python
BoxLjung
````

+ __Box-Pierce Test__

````python
BoxPierce=sm.stats.diagnostic.acorr_ljungbox(residuals.dropna(),boxpierce=True)
BoxPierce
````

## Regression Models

Simple regression model

````python
trend=stl.trend
trend.columns=["trend"]
model=sm.OLS(data,trend,missing="drop")
model=model.fit()
plt.plot(data)
plt.plot(model.fittedvalues)

plt.show()
````

The model with trend and seasonality

````python
season=stl.seasonal
season.columns=["season"]
exog=pd.concat([trend,np.square(trend),season],axis=1)
model=sm.OLS(data,exog,missing="drop")
model=model.fit()
model.summary()
````

The sinusoidal model

````python
exog=pd.concat([trend,np.square(trend),np.sin(2*np.pi*trend/12),np.cos(2*np.pi*trend/12)],axis=1)
model=sm.OLS(data,exog,missing="drop")
model=model.fit()
model.summary()
````
